```python
import re
import requests
import pandas as pd
import numpy as np
```

```python
article_no = 236

ctext = ''

for i in np.arange(0,article_no,100):
    text = str(requests.get('https://search.api.cnn.io/content?size=10000&q=cryptocurrency%20%7C%20cryptocurrencies&from={}'.format(i)).text)
    ctext += text
```

```python
stext = ctext.split('"')
```

```python
print(stext.count('headline'))
print(stext.count('firstPublishDate'))
print(stext.count('body'))
```

```python
headlines = []
publish = []
body = []

j = 0

for i in stext:
    if i == 'headline':
        headlines.append(stext[j+2])
    if i == 'firstPublishDate':
        publish.append(stext[j+2])
    if i == 'body':
        body.append(stext[j+2])
    j += 1
```

```python
scraped_news = pd.DataFrame({'headline' : headlines, 'publish_date' : publish, 'text' : body}).sort_values(by = 'publish_date')
scraped_news
```
